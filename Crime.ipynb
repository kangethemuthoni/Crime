```python
data_df = data_cleaned.head(100000)
```

```python
#check for missing values
data_df.isnull().sum()
```

```python
#function to fill missing Mocodes based on Crime Code 1 mode
def fill_missing_mocodes(row, data):
    if pd.isna(row['Modus Operandi Codes']):
        mode_values = data[data['Crime Code'] == row['Crime Code']]['Modus Operandi Codes'].mode()
        if not mode_values.empty:
            return mode_values[0]
    return row['Modus Operandi Codes']
```

```python
#apply the function to fill missing Mocodes
data_df.loc[:, 'Modus Operandi Codes'] = data_df.apply(lambda row: fill_missing_mocodes(row, data_df), axis=1)
```

```python
#check for remaining missing values in Modus Operandi Codes
missing_mocodes_count = data_df['Modus Operandi Codes'].isna().sum()
print(f"Remaining missing Modus Operandi Codes: {missing_mocodes_count}")
```

```python
data_df.loc[:, 'Modus Operandi Codes'] = data_df['Modus Operandi Codes'].fillna('0')
```

```python
data_df['Modus Operandi Codes'].isna().sum()
```

```python
data_df.head(30)
```

```python
#check the relationship between Weapon Used Code and Crime Code
weapon_crime_relation = data_df.groupby(['Crime Code', 'Weapon Used Code']).size().reset_index(name='count')
print(weapon_crime_relation.sort_values(by='count', ascending=False).head(10))
```

```python
#function to fill missing Weapon Used Codes based on Crime Code mode
def fill_missing_weapon(row, data):
    if pd.isna(row['Weapon Used Code']):
        mode_values = data[data['Crime Code'] == row['Crime Code']]['Weapon Used Code'].mode()
        if not mode_values.empty:
            return mode_values[0]
    return row['Weapon Used Code']
```

```python
#apply the function to fill missing Weapon Used Codes
data_df.loc[:, 'Weapon Used Code'] = data_df.apply(lambda row: fill_missing_weapon(row, data_df), axis=1)
```

```python
#check for remaining missing values in Weapon Used Codes
missing_weapon_count = data_df['Weapon Used Code'].isna().sum()
print(f"Remaining missing Weapon Used Codes: {missing_weapon_count}")
```

```python
data_df.loc[:, 'Weapon Used Code'] = data_df['Weapon Used Code'].fillna(0)
```

```python
#check for missing values
data_df.isnull().sum()
```

```python
data_df.loc[:, 'Victim Sex'] = data_df['Victim Sex'].ffill()
```

```python
data_df.loc[:, 'Victim Descent'] = data_df['Victim Descent'].ffill()
```

```python
#calculate the percentage of similarity between the columns Crime Code and Crime Code 1
def calculate_similarity(col1, col2):
    #ensure both columns are of the same length
    min_length = min(len(col1), len(col2))
    #compute the number of matching values
    matches = (col1[:min_length] == col2[:min_length]).sum()
    #calculate the similarity percentage
    similarity_percentage = (matches / min_length) * 100
    return similarity_percentage

similarity = calculate_similarity(data_df['Crime Code'], data_df['Crime Code 1'])
print(f"Similarity Percentage: {similarity:.2f}%")
```

```python
maindata = data_df.drop('Crime Code 1', axis=1)
```

```python
#function to fill missing Weapon Description based on Weapon Used Code
def fill_missing_weapond(row, data):
    if pd.isna(row['Weapon Description']):
        mode_values = data[data['Weapon Used Code'] == row['Weapon Used Code']]['Weapon Description'].mode()
        if not mode_values.empty:
            return mode_values[0]
    return row['Weapon Description']
```

```python
#apply the function to fill missing Weapon Descriptions
maindata.loc[:, 'Weapon Description'] = maindata.apply(lambda row: fill_missing_weapond(row, data_df), axis=1)
```

```python
#horizontal bar graph of crime distribution by area

plt.figure(figsize=(14, 10))
sns.countplot(y='Area Name', data=maindata, order=maindata['Area Name'].value_counts().index)
plt.title('Crime Distribution by Area')
plt.xlabel('Number of Crimes')
plt.ylabel('Area Name')
plt.show()
```

```python
#vertical bar graph of top 10 crime types

#calculate the top 10 crime types
top_crimes = maindata['Crime Code Description'].value_counts().head(10)

#plot the vertical bar graph
plt.figure(figsize=(12, 8))
sns.barplot(x=top_crimes.index, y=top_crimes.values)
plt.title('Top 10 Crime Types')
plt.xlabel('Crime Type')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#histogram of victim ages

plt.figure(figsize=(12, 6))
sns.histplot(maindata['Victim Age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Victim Ages')
plt.xlabel('Age')
plt.ylabel('Number of Victims')
plt.show()
```

```python
#scatter plot of victim age against top 10 crime types

#calculate the top 10 crime types
top_crimes = maindata['Crime Code Description'].value_counts().head(10).index

#filter data to include only top 10 crimes
scatter_data = maindata[maindata['Crime Code Description'].isin(top_crimes)].dropna(subset=['Victim Age'])

#plotting
plt.figure(figsize=(14, 8))
sns.scatterplot(data=scatter_data, x='Victim Age', y='Crime Code Description', hue='Crime Code Description', alpha=0.6, palette='Set1', legend=None)
plt.title('Scatter Plot of Victim Age Against Top 10 Crime Types')
plt.xlabel('Victim Age')
plt.ylabel('Crime Type')
plt.show()
```

```python
#heatmap of top 20 crime types by area

#calculate the top 20 crime types
top_20_crimes = maindata['Crime Code Description'].value_counts().head(20).index

#filter data to include only the top 20 crimes
filtered_data = maindata[maindata['Crime Code Description'].isin(top_20_crimes)]

#calculate the count of each crime type in each area
area_crime_counts = filtered_data.pivot_table(index='Area Name', columns='Crime Code Description', aggfunc='size', fill_value=0)

#plotting
plt.figure(figsize=(16, 12))
sns.heatmap(area_crime_counts, cmap='Reds', annot=True, fmt='d')
plt.title('Heatmap of Top 20 Crime Types by Area')
plt.xlabel('Crime Type')
plt.ylabel('Area')
plt.xticks(rotation=45, ha='right')
plt.show()
```
